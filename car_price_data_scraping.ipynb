```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.brecorder.com/news/40111873"
```

```python
r = requests.get(url)
htmlContent = r.content
# print(htmlContent)
```

```python
soup = BeautifulSoup(htmlContent, "html.parser")
# print(soup.prettify)
```

```python
title = soup.title
```

```python
paras = soup.find_all("p")
# print(paras)
```

```python
anchors = soup.find_all("a")
# print(anchors)
```

```python
print(soup.find("p"))
```

```python
print(soup.find("strong"))
```

```python
print(soup.find("p").get_text())
```

```python
main_code = soup.find("code").get_text()
print(main_code)
```

```python
import pandas as pd
from io import StringIO
import re
df = pd.read_csv(StringIO(re.sub(r'[=-]','',main_code)), sep='\s{2,}', engine="python")
df
```

```python
del(df["Prices"])
```

```python
df
```

```python
df = df.reset_index()
```

```python
df.dropna(inplace=True)
```

```python
df["Fully"].dtype
```

```python
df["Fully"] = df["Fully"].str.replace("/","")
```

```python
df["Product Description"] = df["Product Description"].str.replace("?"," ")
```

```python
df.columns = ["Car Name", "Price"]
```

```python
df = df.iloc[2:]
```

```python
df["Price"] = df["Price"].str.replace(",","")
```

```python
df=df.reset_index(drop=True)
```

```python
df["Price"] = pd.to_numeric(df["Price"])
```
